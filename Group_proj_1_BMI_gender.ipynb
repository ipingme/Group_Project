```python
# Imobport Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Reference Physical activity data
csv_path = "Resources/cleaned_BMI_data.csv"

# Import the Table_of_Physical_Activity.csv file as a DataFrame
BMI_df = pd.read_csv(csv_path)
BMI_df.head(200)
```

```python
# Narrowed DF to only show certain columns

narrowed_df = BMI_df[["Entity", "Year", "Men Overweight or Obese (%)", "Women overweight or obese (%)", "Men (BMI)", "Women (BMI)"]]
narrowed_df.head(11)
```

```python
# Limit list to 2013 year only
summary_list = narrowed_df[narrowed_df["Year"]==2013]

summary_list.head(11)
```

```python
# Semi-random country sample data set comparing US to semi-random sample of world

AngolaVals = summary_list["Entity"] =="Angola"
EstoniaVals = summary_list["Entity"] =="Estonia"
FijiVals = summary_list["Entity"] =="Fiji"
LesothoVals = summary_list["Entity"] =="Lesotho"
LuxembourgVals = summary_list["Entity"] =="Luxembourg"
NorthKoreaVals = summary_list["Entity"] =="North Korea"
RomaniaVals = summary_list["Entity"] =="Romania"
TurkmenistanVals = summary_list["Entity"] =="Turkmenistan"
UzbekistanVals = summary_list["Entity"] =="Uzbekistan"
VenezuelaVals = summary_list["Entity"] =="Venezuela"
UnitedStatesVals = summary_list["Entity"] =="United States"


summary_list = summary_list[(AngolaVals) | (EstoniaVals) | (FijiVals) | (LesothoVals) | (LuxembourgVals) | (NorthKoreaVals) | (RomaniaVals) | (TurkmenistanVals) | (UzbekistanVals) | (VenezuelaVals)  | (UnitedStatesVals) ]
summary_list.head(5000)
```

```python
# Set x axis and tick locations

x_axis = np.arange(len(summary_list))
tick_locations = [value+0.4 for value in x_axis]
```

```python
# Create a list showing where to write x labels
plt.figure(figsize=(20,7))
plt.bar(x_axis, summary_list["Men Overweight or Obese (%)"], color='b', alpha=0.3, align="edge")
plt.xticks(tick_locations, summary_list["Entity"], rotation="vertical")

plt.rc('font', size = 20)

# Set a Title and labels
plt.title("Men % Overweight or Obese (US vs. Sample Rest of World)")
plt.xlabel("Country")
plt.ylabel("%")
#ax.set_title(plt.title, fontdict={'fontsize': 20}, loc='center')

# Set x and y limits; horizontal line at US
plt.xlim(-0.25, len(x_axis))
plt.ylim(0, max(summary_list["Men Overweight or Obese (%)"])+20)

plt.axhline(y=72.3,linestyle='--',color='red', xmin=-0.5)

plt.savefig("Output/men_overweight_world_sample.png")
```

```python
# Create a list showing where to write x labels
plt.figure(figsize=(20,7))
plt.bar(x_axis, summary_list["Women overweight or obese (%)"], color='m', alpha=0.3, align="edge")
plt.xticks(tick_locations, summary_list["Entity"], rotation="vertical")

plt.rc('font', size = 20)

# Set a Title and labels
plt.title("Women % Overweight or Obese (US vs. Sample Rest of World)")
plt.xlabel("Country")
plt.ylabel("%")

# Set x and y limits; horizontal line at US
plt.xlim(-0.25, len(x_axis))
plt.ylim(0, max(summary_list["Women overweight or obese (%)"])+20)

plt.axhline(y=62.4,linestyle='--',color='red', xmin=-0.5)

plt.savefig("Output/women_overweight_world_sample.png")
```

```python
# Create a list showing where to write x labels
plt.figure(figsize=(20,7))
plt.bar(x_axis, summary_list["Men (BMI)"], color='b', alpha=0.3, align="edge")
plt.xticks(tick_locations, summary_list["Entity"], rotation="vertical")

plt.rc('font', size = 20)

# Set a Title and labels
plt.title("Men BMI (US vs. Sample Rest of World)")
plt.xlabel("Country")
plt.ylabel("BMI")

# Set x and y limits; horizontal line at US
plt.xlim(-0.25, len(x_axis))
plt.ylim(10, max(summary_list["Men (BMI)"])+5)

plt.axhline(y=28.8,linestyle='--',color='red', xmin=-0.5)

plt.savefig("Output/men_bmi_world_sample.png")
```
